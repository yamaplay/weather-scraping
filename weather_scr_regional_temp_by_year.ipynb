```python
import matplotlib.pylab as plt
import pandas as pd
from matplotlib import dates as mdates

plt.rcParams['font.family'] = 'MS Gothic'

df = pd.read_csv('.csv', parse_dates=[''], index_col='', encoding = "SHIFTJIS")
for year in range(1980, 2021, 10):
    h1, = plt.plot(df[df.index.year==year]['()'],label=f'{year}/1~{year}/12')
    plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%#m'))
    plt.legend(handles=[h1])
    plt.show()
    
```

```python
plt.rcParams['font.family'] = 'MS Gothic'

df = pd.read_csv('.csv', parse_dates=[''], index_col=''
                 , encoding="SHIFTJIS")
fig, axes = plt.subplots(3, 2, figsize=(10, 10))
axes = axes.flatten()
years = range(1980, 2021, 10)
for year, ax in zip(years, axes):
  df[df.index.year==year].plot(
    y='()', x_compat=True, xlabel='',
    label=f'{year}/1~{year}/12', title=f'{year} ()', ax=ax)
  ax.set_ylim([0., 32.])
  ax.xaxis.set_tick_params(reset=True)
  ax.xaxis.set_major_locator(mdates.MonthLocator())#X
  ax.xaxis.set_major_formatter(mdates.DateFormatter('%#m'))
  ax.grid(axis='y')
  ax.tick_params(top=False)
  ax.legend(loc='upper left')
  ax.set_frame_on(False)

# 
axes[-1].axis('off')

fig.tight_layout(pad=2.0)

plt.show()
```
